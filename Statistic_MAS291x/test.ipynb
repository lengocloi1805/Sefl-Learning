```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

### Linear with one variation

```python
# Linear equation with one variation unknown
X = np.array([10, 16, 18, 24, 48, 56])
y = np.array([12, 14, 13, 18, 38, 64])
X = X.reshape(-1,1) #if we don't reshape, we can use np.array([[1],[1.5],[2],[2.5],[3],[3.5]])
```

```python
regression_model = LinearRegression().fit(X, y)
```

```python
b = regression_model.coef_
a = regression_model.intercept_
print('y = ' +  str(a)+' + '+str(*b) + 'x')
```

```python
pred = a + b*18
pred
```

### Mean and Std for grouped data

https://stats.libretexts.org/Learning_Objects/02%3A_Interactive_Statistics/08%3A_Mean_and_Standard_Deviation_for_Grouped_Frequency_Tables_Calculator

```python
x = [(10,10.5), (10.5,11), (11,11.5), (11.5,12), (12,12.5), (12.5,13), (13,13.5)]
y = [4, 12, 26, 37, 43, 28, 10]
```

```python
def mean_std(x,y):
    n = sum(y)
    mid = [(u+v)/2 for u,v in x]
    sum_f1 = sum([mid[i]*y[i] for i in range(len(y))])s
    mean = sum_f1/n
    sum_f2 = sum([y[i]*((mid[i]-mean)**2) for i in range(len(y))])

    var = sum_f2/(n-1) #if population, we use N instead n-1
    std = var**.5
    
    return mean, var, std
```

```python
mean, var, std = mean_std(x,y)
print(f'mean:{mean:.4f}, var:{var:.4f}, std:{std:.4f}')
```
